```python
from gensim.models import Word2Vec
import gensim

f1 = open('Harry_Potter_Camara_Secreta-br.txt', encoding='UTF8')
f2 = open('Harry_Potter_e_A_Pedra_Filosofal.txt', encoding='UTF8')

sents = []

for line in f1:
    tokens = list(gensim.utils.tokenize(line, lower=True))
    sents.append(tokens)

for line in f2:
    tokens = list(gensim.utils.tokenize(line, lower=True))
    sents.append(tokens)

print(sents)
```

```python
model = Word2Vec(sents, vector_size=300, epochs=20)

model.wv['harry']
print(model.wv.most_similar('harry'))
#model.wv.most_similar(positive=[...], negative=[...])
model.wv.similarity('harry', 'hagrid')
model.wv.similarity('harry', 'dobby')

model.wv.save_word2vec_format('model_harry.txt', binary=False)

f1.close()
f2.close()

#python -m gensim.scripts.word2vec2tensor -i model_harry.txt -o model_harry
#https://projector.tensorflow.org
```
